```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'


from matplotlib import pyplot as plt

import numpy as np
```

```python
from sigmoid import σ
```

```python
xs = np.linspace(-5, 5, 20)
α = σ(xs)

plt.plot(xs, α, 'r.-')
plt.title('sigmoid function')
plt.ylabel('activation')
plt.xlabel('input')
plt.show()
```
